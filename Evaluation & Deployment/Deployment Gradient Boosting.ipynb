```python
# Haalt de Gradient Boosting classifier op uit de lokale map.
with open('pipelineGB.pkl', 'rb') as file:
    gradient_boosting_pipe = pickle.load(file)
```

```python
# Uit de gridsearch zijn deze parameters als de beste samenstelling gekomen en daarom dienen zij hier aangepast te worden.
gradient_boosting_pipe.n_estimators = 3000
gradient_boosting_pipe.learning_rate = 1
gradient_boosting_pipe.max_depth = 10
```

```python
# Hier wordt de voorspelling gedaan op de data wie wel of niet zouden willen overstappen.
predicted = gradient_boosting_pipe.predict(data_ev)
```

```python
# Hier wordt bepaald welke klanten zouden willen overstappen.
rate = []

for i in range(len(predicted)):
    if (predicted[i] == 1):
        rate.append([str((i + 1))])
```

```python
# Hier wordt tarief2.csv aangemaakt met de klanten daarin de klanten die zouden willen overstappen.
with open('tarief2.csv', 'w', newline = '') as f:
    writer = csv.writer(f)
    writer.writerows(rate)
```

```python
# Hier wordt de winstverwachting uitgerekend en getoond op de klanten die zouden willen overstappen.
W = (((data_ev['Seconds of Use'] / 60.0) * 0.2) + (data_ev['Frequency of SMS'] * 0.07))
expected_profit = (1 - (0.25 * predicted)) * W
print("De winstverwachting is: €", (round(expected_profit.sum(), 2)), sep = "", end = "0.")
```
