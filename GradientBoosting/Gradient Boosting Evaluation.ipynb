```python
# Haalt de gradient boosting op uit de lokale map.
with open('pipelineGradientBoosting.pkl', 'rb') as file:
    gradient_boosting_pipe = pickle.load(file)
```

```python
# Hier wordt het voorspellen op de testdata gedaan.
pred = gradient_boosting_pipe.predict(X_test)
```

```python
# Confusion matrix
ConfusionMatrixDisplay.from_predictions(y_test, pred, cmap = 'Blues')
plt.show()
```

```python
# Hier worden de accuracy, f1, precision en recall scores berekend en gegeven.
acc = accuracy_score(y_test, pred)
rec = recall_score(y_test, pred)
pre = precision_score(y_test, pred)
fsco = f1_score(y_test, pred)
print("Gradient Boosting Classifier accuracy is: {:.2f}".format(acc))
print("Gradient Boosting Classifier recall is: {:.2f}".format(rec))
print("Gradient Boosting Classifier precision is: {:.2f}".format(pre))
print("Gradient Boosting Classifier F1 score is: {:.2f}".format(fsco))
```
