```python
#Parameters voor de gradient boosting algoritme.
params = {
    "n_estimators": 2000,
    "max_depth": 10,
    "min_samples_split": 5,
    "learning_rate": 0.01,
    "loss": "log_loss",
}
```

```python
# De churn wordt uit de traindata gelaten.
train = data.drop(columns = ['Churn'])
Y = data['Churn']

# Het splitten van de data wordt hier gedaan.
X_train, X_test, y_train, y_test = train_test_split(train, Y, test_size = 0.2, random_state = 1987)
```

```python
# Hier wordt het fitten gedaan op de data.
# We passen hier een pipeline toe voor de traindata door middel van Gradient Boosting.
pipeline = Pipeline([
    ('model', ensemble.GradientBoostingClassifier(**params))
])

pipeline.fit(X_train, y_train)
```

```python
# We slaan de pipeline op als een pickle bestand.
with open('pipelineGradientBoosting.pkl', 'wb') as file:
    pickle.dump(pipeline, file)
```
